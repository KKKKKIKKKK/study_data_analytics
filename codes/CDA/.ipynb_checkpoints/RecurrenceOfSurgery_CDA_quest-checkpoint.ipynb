#### quest
- 선정 변수 : 체중, 재발여부, ?(범주형)
- 연속형 / 범주형(정규 / 비정규 분포) 검증
- 분석 스토리 작성
- ex) 재발 여부가 정규 분포이면 비정규 분포인 범주형 변수 찾고 검증
#### 분석 스토리
- 체중이 재발여부와 직업에 미치는 영향이 있는지 확인하고자 한다.
- pvalue=2.67e-132 값을 근거로 체중에 따라 재발여부의 평균의 차이가 있음을 확인하였다.
- pvalue=0.57 값을 근거로 직업에 따라 (무직, 학생) 평균에 차이가 없음을 확인하였다.
- 분석결론
    - 체중은 척주질병의 재발여부에 영향을 끼치며, 직업에는 별다른 영향을 끼치지 않음을 알수 있다.

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rc('font',family='Malgun Gothic')
import seaborn as sns
import pandas as pd
import scipy.stats as stats
```

```python
df_ROS = pd.read_csv('../../datasets/RecurrenceOfSurgery.csv')
df_ROS.head(2)
```

#### Y:연속형 : 단일 변수에 대한 정규성 여부 판단
- 검증 갯수 따른 방법 선택 (레코드 갯수)
    - 5천개보다 이하 : shapiro normality test
    - 5천개보다 이상 : Anderson normality test

#### 선정 변수: 체중
- 분석 내용 : 체중이 정규 분포를 따르는가?
    - 귀무가설 : 정규 분포이다.
    - 대립가설 : 정규 분포가 아니다.
- 분석결론 (기준:p-value 0.05) / P.value < 0.05 : 대립 참
    - 통계 결론 : pvalue=8.31e-16이므로 대립가설 참
    - 사실 결론 : 체중은 정규 분포가 아니다

```python
df_weight = df_ROS['체중']
```

```python
sns.histplot(data=df_ROS, x='체중', kde=True)
plt.show()
```

```python
df_weight.count() # 5천개 이하 -> shapiro
```

```python
stats.shapiro(df_weight)

## pvalue=8.31e-16 ->0.05보다 작음, 대립가설 참, 비정규분포이다.
```

### 범주형 / 연속형 : 범주형 따른 상관관계 확인
#### 선정 변수: 재발여부(항목 2), 체중
- 분석 내용 : 체중에 따라 재발여부의 평균의 차이가 있는가?
    - 귀무가설 : 평균에 차이 없다.
    - 대립가설 : 평균에 차이 있다.
- 분석결론 (기준:p-value 0.05) / P.value < 0.05 : 대립 참
    - 통계 결론 : pvalue=2.67e-132, 대립가설 참
    - 사실 결론 : 평균에 차이 있다.

```python
df_ROS.columns
```

```python
sns.histplot(data=df_ROS, x='체중')
```

```python
df_ROS['체중'].value_counts() 
```

```python
df_condition = df_ROS[['체중','재발여부']]
df_condition
```

```python
df_condition['체중'].value_counts()
```

#### 정규성 검증

```python
stats.shapiro(df_condition['재발여부'])

# pvalue=0.0 -> 0.05보다 작음 /비정상/ 대립가설 참/비정규분포
```

```python
sns.histplot(df_condition, x='체중', kde=True)
plt.show()
```

### 범주형 항목에 따라 검증
- 평균의 차가 있느냐 없느냐를 확인
- 결론 : 평균의 차이가 없다.

```python
df_condition['재발여부'].value_counts()
```

```python
df_weight_0 = df_condition[df_condition['재발여부'] == 0]
df_weight_0[:2]
```

```python
df_weight_1 = df_condition[df_condition['재발여부'] == 1]
df_weight_1[:2]
```

### 검증
- pvalue=2.67e-132 -> 0.05보다 작음/비정상/차이가 있다.

```python
stats.ranksums(df_weight_0['재발여부'],df_weight_1['재발여부'] )
```

```python
sns.histplot(data=df_condition, x='체중', kde=True, hue='재발여부')
plt.show()
```

### 선정 변수: 체중(항목 2), 직업[:100](정규분포)
- 분석 내용 : 직업에 따른 체중 평균의 차이가 있는가?
    - 귀무가설 : 평균에 차이 없다.
    - 대립가설 : 평균에 차이 있다.
- 분석결론 (기준:p-value 0.05) / P.value < 0.05 : 대립 참
    - 통계 결론 : pvalue=0.57, 귀무가설 참
    - 사실 결론 : 평균에 차이 없다./정규분포/등분산

```python
sns.histplot(data=df_ROS, x='체중', kde=True, hue='직업')
```

```python
df_ROS['직업'].value_counts() #무직, 학생
```

```python
## 검증 대상 항목만 필터링 and 2000개만 추출(정규분포 자료 획득하기 위함)
condition = (df_ROS['직업'] == '무직') |(df_ROS['직업'] == '학생')
df_condition_second = df_ROS[condition][['체중','직업']][:100]
df_condition_second[:2]
```

```python
df_condition_second.isnull().sum()
```

```python
df_condition_second['직업'].value_counts()
```

### 정규성 검증
- 5천개 이하 -> shapiro

```python
stats.shapiro(df_condition_second['체중'])

# pvalue=0.0668 -> 0.05보다 큼/정상/귀무가설참/정규분포
```

```python
sns.histplot(data=df_condition_second, x='체중', kde=True)
plt.show()
```

#### 등분산 검증

```python
sns.countplot(data=df_condition_second, x='직업')
plt.show()
```

```python
df_no = df_condition_second[df_condition_second['직업'] == '무직']
df_no[:2]
```

```python
df_student = df_condition_second[df_condition_second['직업'] == '학생']
df_student[:2]
```

### 등분산 검증
- pvalue=0.92 -> 0.05보다 큼, 귀무가설 참, 분산이 같다. 등분산

```python
stats.levene(df_no['체중'], df_student['체중'])
```

```python
# 시각화
sns.histplot(data=df_condition_second, x='체중', kde=True, hue='직업')
plt.show()
```

### 범주형이 2개임 -> t-test

- pvalue=0.57 귀무가설 참

```python
stats.ttest_ind(df_no['체중'], df_student['체중'],equal_var=True )
```
