```python
import pymongo as mg
import pandas as pd
```

```python
mg.__version__
```

```python
client = mg.MongoClient(host='mongodb://localhost:27017')
```

```python
database = client['db_NHIS']
```

```python
collection = database['NSC2_BND']
```

```python
cursor = collection.find({})
```

```python
list_BND = list(cursor) # cursor 맨처음라인 리스트 동작시키면 끝(end)로 감.
# 데이터가 없어져서, 다시 진행위해 데이터 다시(cursor find)호출 
```

```python
list_BND
```

```python
df_BND = pd.DataFrame(list_BND)
```

```python
df_BND
```

```python
df_BND.to_csv('./NSC2_BND.csv') #현재폴더에, 데이터베이스 이름..csv 경로에 넣기.  
```
