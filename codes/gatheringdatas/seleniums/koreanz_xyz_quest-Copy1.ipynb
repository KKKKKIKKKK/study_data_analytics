###  드라마 페이지 정보 
- 게시물번호 tag 
    -  div.wr-num.hidden-xs
- 게시물 제목 tag 
    -  div.wr-subject
- 게시물 날짜 tag 
    - div.wr-date.hidden-xs
- 게시물 조회수 
    - div.wr-hit.hidden-xs

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

numbers = []
subjects = []
dates = []
views = []
links = []

for i in range(1, 11):  # 3페이지까지 url 가져오기
    browser.get('https://a73.koreanz.xyz/bbs/board.php?bo_table=tvdrama&page={0}'.format(i))

    while True:
        bh = browser.execute_script("return document.body.scrollHeight")  # 브라우저 상의 처음 높이
        print(bh)
        time.sleep(4)
        browser.execute_script("window.scrollTo(0, document.body.scrollHeight)")  # 스크롤 내리기
        time.sleep(2)
        ah = browser.execute_script("return document.body.scrollHeight")  # 브라우저 상의 처음 높이
        if ah == bh:
            break

    infos = browser.find_elements(By.CSS_SELECTOR, ".list-item")  # 전체
    for info in infos:
        number = info.find_element(By.CSS_SELECTOR, '.wr-num.hidden-xs').text
        numbers.append(number)
        subject = info.find_element(By.CSS_SELECTOR, '.wr-subject').text
        subjects.append(subject)
        date = info.find_element(By.CSS_SELECTOR, '.wr-date.hidden-xs').text
        dates.append(date)
        view = info.find_element(By.CSS_SELECTOR, '.wr-hit.hidden-xs').text
        views.append(view)
        link = info.find_element(By.CSS_SELECTOR, 'a.item-subject').get_attribute("href")
        links.append(link)
# 웹 브라우저를 닫습니다.
browser.close()

# 데이터를 출력하여 확인
for i in range(len(numbers)):
    print(f"Number: {numbers[i]}, Subject: {subjects[i]}, Date: {dates[i]}, View: {views[i]}, Link: {links[i]}")
```

```python
# 데이터프레임을 생성합니다.
df = pd.DataFrame({'number': numbers, 'subject': subjects, 'date': dates, 'view': views, 'link': links})
print(df)
```
