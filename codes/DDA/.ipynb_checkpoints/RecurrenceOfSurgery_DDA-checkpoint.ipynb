## quest2
- 직업군이 상위 2위에 속하는 항목 찾기. valuecount로 찾기 (빈도수나옴)
- 해당 결괏값을 데이터 프레임으로 만들기 /슬라이싱으로 사용 가능 
- ※직업군이 상위 2위에 속하는 항목 :value_count, DataFrame, iloc
- ※체중> 70,  신장> 170, 흡연여부 없는 분
- 해당 레코드 갯수 표현 

```python
df_ROSjob = df_ROS_forpatient['직업'].value_counts().iloc[:2]
df_ROSjob
```

```python
# condition = (df_ROS_forpatient['직업'] == '사무직') | (df_ROS_forpatient['직업'] == '자영업') 
condition = (df_ROS_forpatient['직업'] == df_ROSjob.index[0]) | (df_ROS_forpatient['직업'] == df_ROSjob.index[1])
condition2 = (df_ROS_forpatient['체중'] > 70) & (df_ROS_forpatient['신장'] > 170) & (df_ROS_forpatient['흡연여부'] == 0) 
```

```python
condition
```

```python
df_RODquest2 = df_ROS_forpatient.loc[condition & condition2]
```

```python
df_RODquest2
```

```python
df_RODquest2['직업'].value_counts()
```

```python
df_RODquest2['흡연여부'].value_counts()
```

```python
df_RODquest2['신장'].value_counts()
```

```python
df_RODquest2['체중'].value_counts()
```
